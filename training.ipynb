```python
import tensorflow as tf
from google.protobuf import text_format
from object_detection.protos import pipeline_pb2
from object_detection.utils import config_util
```

```python
cfg = config_util.get_configs_from_pipeline_file(FILES['PIPELINE_CFG'])
pipeline_cfg = pipeline_pb2.TrainEvalPipelineConfig()
```

```python
with tf.io.gfile.GFile(FILES['PIPELINE_CFG'], "r") as f:                                                                                                                                                                                                                     
    proto_str = f.read()                                                                                                                                                                                                                                          
    text_format.Merge(proto_str, pipeline_cfg)  
```

```python
config_text = text_format.MessageToString(pipeline_cfg)                                                                                                                                                                                                        
with tf.io.gfile.GFile(FILES['PIPELINE_CFG'], "wb") as f:                                                                                                                                                                                                                     
    f.write(config_text)  
```
